```python
#toplulaştırma ve gruplama
```

```python
df["age"].mean() #yaş ortalaması
```

```python
df.groupby("sex")["age"].mean() #erkeklerin ve kadınların yaş ortalaması

#df'i cinsiyete göre grupla sonra yaş değişkeninin ortalamasını al demektir
```

```python
df.groupby("sex").agg({"age": "mean"}) 

#bu kullanım daha iyi
```

```python
df.groupby("sex").agg({"age": ["mean", "sum"]})

#yaşın ortalaması ve toplamı
```

```python
df.groupby("sex").agg({"age": "mean",
                       "survived": "mean"})

#sex'e göre grupla sonra cinsiyete göre yaş ortalamasını al ve cinsiyete göre survived ortalamasını al
```

```python
df.groupby(["sex", "embark_town"]).agg({"age": ["mean"],
                       "survived": "mean"})
```

```python
df.groupby(["sex", "embark_town", "class"]).agg({"age": ["mean"],
                       "survived": "mean"})
```

```python
df.groupby(["sex", "embark_town", "class"]).agg({
    "age": ["mean"],
    "survived": "mean",
    "sex": "count"})
```

```python
#merge ile birleştirme yöntemleri

df1 = pd.DataFrame({'employees': ['john', 'dennis', 'mark', 'maria'],
                    'group': ['accounting', 'engineering', 'engineering', 'hr']})
df1
```

```python
df2 = pd.DataFrame({'employees': ['mark', 'john', 'dennis', 'maria'],
                    'start_date': [2010, 2009, 2014, 2019]})
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, on="employees")
```

```python
# Amaç: Her çalışanın müdürünün bilgisine erişmek istiyoruz.
```

```python
df3 = pd.merge(df1, df2)
df3
```
